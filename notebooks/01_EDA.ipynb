# RetailPulse - Exploratory Data Analysis (EDA)
This notebook covers Week 1 objectives: data exploration, distributions, and correlation analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set style
sns.set_theme(style="whitegrid")

base_dir = os.path.dirname(os.getcwd())
sales_path = os.path.join(base_dir, 'data', 'raw', 'sales_data.csv')
sales_df = pd.read_csv(sales_path)
print(sales_df.head())
```

### Sales Distribution Over Time

```python
sales_df['Date'] = pd.to_datetime(sales_df['Date'])
daily_sales = sales_df.groupby('Date')['TotalAmount'].sum().reset_index()

plt.figure(figsize=(15, 6))
sns.lineplot(data=daily_sales, x='Date', y='TotalAmount')
plt.title('Daily Revenue Over Time')
plt.show()
```
